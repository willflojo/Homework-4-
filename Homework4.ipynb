```python
import numpy as np 
import matplotlib.pyplot as plt

marks = {'Andy':88,'Amy':66,'James':90,'Jules':55,'Arthur':77}
names = []
scores = []

for key in marks:
    names.append(key)
    scores.append(marks[key])


plt.bar(names,scores)
plt.title("Student Scores")
plt.show()

plt.pie(scores,labels=names)
plt.title("Student Scores")
plt.legend(title="Student Names", bbox_to_anchor=(1.05, 1.0),loc = 'upper left')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
sales = np.array([9,7,8,3,7])
expenses = np.array([3,2,2,1,3])
revenue = np.array([7,5,6,2,5])
traffic = np.array([12,15,10,13,10,12,13,13,13,14,12])
items = np.array(['Books','Paper','Pens','Pencils'])
numitems = np.array([10,20,15,5])

plt.subplot(2,3,1)
plt.plot(x,sales,x,revenue)
plt.title("Revenue and Sales")

plt.subplot(2,3,2)
plt.bar(x,sales)
plt.title("Sales per day")

plt.subplot(2,3,3)
plt.pie(numitems,labels = items)
plt.title("Item Sales")

plt.subplot(2,3,4)
plt.plot(x,expenses)
plt.title("Expenses per day")
plt.grid()

plt.subplot(2,3,5)
plt.scatter(x,revenue)
plt.title("Revenue per day")

plt.subplot(2,3,6)
plt.hist(traffic)
plt.title("site traffic")
plt.show()
```
